##구매유형 비율 - 저가/중저가/중고가/고가 추가

```python
train_ft['저가_구매비율'] = train_ft['저가_구매횟수'] / train_ft['구매건수']
train_ft['중저가_구매비율'] = train_ft['중저가_구매횟수'] / train_ft['구매건수']
train_ft['중고가_구매비율'] = train_ft['중고가_구매횟수'] / train_ft['구매건수']
train_ft['고가_구매비율'] = train_ft['고가_구매횟수'] / train_ft['구매건수']

test_ft['저가_구매비율'] = test_ft['저가_구매횟수'] / test_ft['구매건수']
test_ft['중저가_구매비율'] = test_ft['중저가_구매횟수'] / test_ft['구매건수']
test_ft['중고가_구매비율'] = test_ft['중고가_구매횟수'] / test_ft['구매건수']
test_ft['고가_구매비율'] = test_ft['고가_구매횟수'] / test_ft['구매건수']
train_ft.shape , test_ft.shape
```

##월초_말_구매건수_변화 (구매 패턴 에측용,,?)
ㄴ 고객의 구매 패턴이 월초에서 월말로 어떻게 변했는지

ㄴ 양수 값 : 월초에 비해 월말에 구매 건수가 증가한 경우. 고객의 구매 패턴이 월말로 갈수록 활성화

ㄴ 음수 값 : 월초에 비해 월말에 구매 건수가 감소한 경우. 고객의 구매가 월말에 덜 활발해졌다는 의미


```python
train_ft['월초_월말_구매건수_변화'] = train_ft['월초_구매건수'] - train_ft['월말_구매건수']

test_ft['월초_월말_구매건수_변화'] = test_ft['월초_구매건수'] - test_ft['월말_구매건수']
train_ft.shape , test_ft.shape
```

## 등급_구매건수 범주화(1-8) 추가
ㄴ 구매건수가 많을수록 1등급

ㄴ train_ft 기준 범주를 test_ft에도 사용 (test_ft에서 따로 계산하면 데이터리퀴지,,,)

구매건수 등급 1 범위: [1.00, 5.00), 개수: 1711

구매건수 등급 2 범위: [5.00, 9.00), 개수: 1747

구매건수 등급 3 범위: [9.00, 14.00), 개수: 1919

구매건수 등급 4 범위: [14.00, 20.00), 개수: 1922

구매건수 등급 5 범위: [20.00, 29.00), 개수: 2002

구매건수 등급 6 범위: [29.00, 42.00), 개수: 1899

구매건수 등급 7 범위: [42.00, 67.00), 개수: 1849

구매건수 등급 8 범위: [67.00, 569.00), 개수: 1890

ㄴ 기존에 있던건 구매 금액별 범주화

```python
train_ft['등급_구매건수'] = pd.qcut(train_ft['구매건수'], q=8, labels=range(8, 0, -1))
test_ft['등급_구매건수'] = pd.qcut(train_ft['구매건수'], q=8, labels=range(8, 0, -1))
train_ft.shape , test_ft.shape
```
