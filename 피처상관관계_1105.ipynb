```python
import pandas as pd
import numpy as np

# 숫자형 피처만 선택하여 상관계수 행렬 계산
numeric_df = train_ft.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

# 상관계수가 양수인 값만 필터링
positive_corr = corr_matrix[(corr_matrix > 0.8) & (corr_matrix < 0.9)].stack()
```

```python
negative_corr = corr_matrix[(corr_matrix < -0.1)].stack()
```
